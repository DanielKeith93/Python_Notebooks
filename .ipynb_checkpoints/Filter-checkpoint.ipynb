```python
fs = 100000.0
lowcut = 300.0
T = 0.018
nsamples = T * fs
from scipy.signal import bessel, lfilter


def bessel_bandpass(lowcut,fs, order=8):
    nyq = 0.5 * fs
    low = lowcut / nyq
    b, a = bessel(order, low, btype='low')
    return b, a


def bessel_bandpass_filter(data, lowcut,fs, order=5):
    b, a = bessel_bandpass(lowcut,fs, order=order)
    y = lfilter(b, a, data)
    return y
```

```python
#define parameters
N_0 = 0.37
N_1 = 1 - N_0
deltat = 0.018 #s, taken from spin to charge optimisation
t_down_in = 0.0039
t_up_out = 0.0069 #Taken from experimental data.
deltaI = 32 * 10 ** - 12 #A
points = nsamples
time = linspace( 0, deltat, points )
a_n = 71 * 10 ** - 15 #A / ( ( Hz ) ^ ( 1 / 2 ) ) #Have to multiply spectral density by a bandwidth.
f_c = 300 #Hz, cut off frequency, also used as the bandwidth
A_n = a_n * ( f_c )
```

```python
from scipy.signal import butter, lfilter

def butter_bandpass(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a

if __name__ == "__main__":
    import numpy as np
    import matplotlib.pyplot as plt
    from scipy.signal import freqz  
    # Sample rate and desired cutoff frequencies (in Hz).
    fs = 1000000
    # Plot the frequency response for a few different orders.
    plt.figure(1)
    plt.clf()
    for order in [2, 4, 6, 8]:
        b, a = butter_bandpass(10000, 20000, fs, order=order)
        w, h = freqz(b, a, worN=2000)#np.logspace(-4, 3, 2000))
        plt.semilogx((fs * 0.5 / np.pi) * w, abs(h), label="order = %d" % order)  
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Gain')
    plt.grid(True)
    plt.legend(loc='best')

    plt.figure(2)
    plt.clf()
    for order in [2, 4, 6, 8]:
        b, a = butter_bandpass(1000, 2000, fs, order=order)
        w, h = freqz(b, a, worN=2000)#np.logspace(-4, 3, 2000))
        plt.semilogx((fs * 0.5 / np.pi) * w, abs(h), label="order = %d" % order)  
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Gain')
    plt.grid(True)
    plt.legend(loc='best')
```
