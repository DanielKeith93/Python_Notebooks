```python
#define parameters
fs = 1000000.0
lowcut = 10000.0
T = 0.055
nsamples = T * fs
N_0 = 0.55
N_1 = 1 - N_0
deltat = T #s, taken from spin to charge optimisation
t_down_in = 0.0051
t_up_out = 0.0065 #Taken from experimental data.
deltaI = 0.45 * 10 ** - 9 #A
points = nsamples
time = linspace( 0, deltat, points )
a_n = 0.24 * 10 ** - 12 #A / ( ( Hz ) ^ ( 1 / 2 ) ) #Have to multiply spectral density by a bandwidth.
f_c = lowcut #Hz, cut off frequency, also used as the bandwidth
A_n = a_n * sqrt( f_c )
bars = 200

from scipy.signal import bessel, lfilter

def bessel_lowpass(lowcut,fs, order=8):
    nyq = 0.5 * fs
    low = lowcut / nyq
    b, a = bessel(order, low, btype='low')
    return b, a


def bessel_lowpass_filter(data, lowcut,fs, order=8):
    b, a = bessel_lowpass(lowcut,fs, order=order)
    y = lfilter(b, a, data)
    return y
```

```python
print A_n
wave = bessel_lowpass_filter( A_n * random.normal( 0, 1, points ), lowcut,fs, order=8)
time = linspace( 0, 1, points )
figure()
plot( time, wave )
figure()
values, bins, _ = hist( wave, 100, histtype = 'step', normed = True )
area = sum(numpy.diff(bins[:])*values[:])
print area
s = np.std( wave )
m = np.mean( wave )
print s, m
```
