```python
#
# Landau-Zener transitions in a quantum two-level system
#
# from qutip.states import *
# from qutip.Qobj import *
# from qutip.tensor import *
# from qutip.ptrace import *
# from qutip.operators import *
# from qutip.expect import *
# from qutip.correlation import *
import qutip as qu

from pylab import *
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
def hamiltonian_t(t, args):
    """ evaluate the hamiltonian at time t. """
    H0 = args[0]
    H1 = args[1]
    return H0 + t * H1

#
# set up the parameters
#
delta = 0.8 * 2 * pi   # qubit sigma_x coefficient
eps0  = 0.0 * 2 * pi   # qubit sigma_z coefficient
A     = 10.0 * 2 * pi   # sweep rate
gamma1 = 0.0 * 2 * pi           # relaxation rate
n_th = 0.0 * 2 * pi             # average number of thermal photons
psi0 = (1)*(1*qu.basis(2,0)+0*qu.basis(2,1))      # initial state

#
# Hamiltonian
#
sx = qu.sigmax()
sy = qu.sigmay()
sz = qu.sigmaz()
sm = qu.destroy(2)

H0 = - delta/2.0 * sx - eps0/2.0 * sz
H1 = - A/2.0 * sz
args = (H0, H1)

#
# collapse operators, only active if gamma1 > 0
#
c_ops = []

rate = gamma1 * (1 + n_th)
if rate > 0.0:
    c_ops.append(sqrt(rate) * sm)       # relaxation

rate = gamma1 * n_th
if rate > 0.0:
    c_ops.append(sqrt(rate) * sm.dag()) # excitation

#
# evolve and calculate expectation values
#
tlist = linspace(-10.0, 10.0, 3000)
output = qu.mesolve(hamiltonian_t, psi0, tlist, c_ops, [sm.dag() * sm], args)
#
# Plot the results
#

plot(tlist, real(output.expect[0]), 'b', tlist, real(1-output.expect[0]), 'r')
plot(tlist, 1 - exp( - pi * delta **2 / (2 * A)) * ones(shape(tlist)), 'k')
xlabel('Time')
ylabel('Occupation probability')
title('Excitation probabilty the two-level system')
legend(("L", "R", "Landau-Zener formula"), loc=0)
show()
```

```python
eigvals = []
for i in tlist:
    H = H0 + H1 * i
    eigvals.append( qu.Qobj.eigenenergies(H) )
plot( tlist, eigvals )
show()
```

```python
#plot everything as a function of detuning

#
# set up the parameters
#
t_c = 0.01 * 2 * pi   # qubit sigma_x coefficient
Delta = 5.0 * 2 * pi   # detuning
time = 50   # sweep time
A = Delta / time   #sweep rate
gamma1 = 0.0005 * 2 * pi           # relaxation rate
n_th = 0.0 * 2 * pi             # average number of thermal photons
psi0 = (1)*(1*qu.basis(2,0)+0*qu.basis(2,1))      # initial state

H0 = - t_c/2.0 * sx
H1 = - A/2.0 * sz
args = (H0, H1)

#
# collapse operators, only active if gamma1 > 0
#
c_ops = []

rate = gamma1 * (1 + n_th)
if rate > 0.0:
    c_ops.append(sqrt(rate) * sm)       # relaxation

rate = gamma1 * n_th
if rate > 0.0:
    c_ops.append(sqrt(rate) * sm.dag()) # excitation

#
# evolve and calculate expectation values
#
dlist = linspace(-Delta, Delta, 1500 )
tlist = linspace(-time, time, 1500)

output = qu.mesolve(hamiltonian_t, psi0, tlist, c_ops, [sm.dag() * sm], args)
#
# Plot the results
#

plot(dlist, real(output.expect[0]), 'b', dlist, real(1-output.expect[0]), 'r')
plot(dlist, 1 - exp( - pi * delta **2 / (2 * A)) * ones(shape(tlist)), 'k')
xlabel('Detuning')
ylabel('Occupation probability')
title('Excitation probabilty the two-level system')
legend(("L", "R", "Landau-Zener formula"), loc=0)
show()
```

```python
def hamiltonian_t(t, args):
    """ evaluate the hamiltonian at time t. """
    H0 = args[0]
    H1 = args[1]
    return H0 + t * H1

#
# set up the parameters
#
dB = 0.8 * 2 * pi   
t_c = 5 * 2 * pi   
v = 10.0 * 2 * pi   # sweep rate

# initial state
psi0 = (0*qu.basis(3,0)+1*qu.basis(3,1)+0*qu.basis(3,2)).unit()      
# print psi0

sm = qu.destroy(3)

H0 = qu.Qobj( [[0,dB/2,0],[dB/2,0,t_c],[0,t_c,0]] )
H1 = qu.Qobj( [[v/2,0,0],[0,v/2,0],[0,0,-v/2]] )
args = (H0, H1)
# print H0+H1

#
# collapse operators
#
c_ops = []


#
# evolve and calculate expectation values
#
tlist = linspace(-10.0, 10.0, 2000)
dlist = linspace(-10.0*v, 10.0*v, 2000)

states = [qu.basis(3,0)*qu.basis(3,0).dag(), qu.basis(3,1)*qu.basis(3,1).dag(), qu.basis(3,2)*qu.basis(3,2).dag()]
# print states

output = qu.mesolve(hamiltonian_t, psi0, tlist, c_ops, states, args)
#
# Plot the results
#

plot(tlist, real(output.expect[0]), 'b', tlist, real(output.expect[1]), 'r', tlist, real(output.expect[2]), 'g')
xlabel('Time (s)')
ylabel('Occupation probability')
title('Excitation probabilty of the three-level system')
legend(("T0", "S11", "S02"), loc=0)
show()
```

```python
eigvals = []
for i in tlist:
    H = H0 + H1 * i
    eigvals.append( qu.Qobj.eigenenergies(H) )
plot( dlist, eigvals )
xlabel('Detuning')
ylabel('Energy')
show()
```
