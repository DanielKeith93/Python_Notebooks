```python
def LZ_pulse(DETUNING, TIME, INIT, TUNNEL, MAG, POINTS ):
    
    dlist = list( np.linspace(DETUNING[0], DETUNING[1], POINTS) )
    if len( DETUNING ) > 2:
        for i in DETUNING[2:]:
            dlist = dlist + list( np.linspace(dlist[-1], i, POINTS) )
    tlist = list( np.linspace(0, TIME[0], POINTS) )
    if len( TIME ) > 1:
        for i in TIME[1:]:
            tlist = tlist + list( np.linspace(tlist[-1], i + tlist[-1], POINTS) )
     
    rates = []
    for i in range(0, len(TIME)):
        rate = (DETUNING[i+1]-DETUNING[i])/TIME[i]
        rates.append( rate )
    
    D_axis = []
    T_axis = []
    def det_pos( t ):
        T_axis.append( t )
        for i in range(0, len(TIME)):
            if t <= sum( TIME[:i+1] ):
                D_axis.append(DETUNING[i] + rates[i] * ( t - sum( TIME[:i] ) ))
                return DETUNING[i] + rates[i] * ( t - sum( TIME[:i] ) )
            if t > sum( TIME ):
                D_axis.append(DETUNING[-1] + rates[-1] * ( t - sum( TIME ) ))
                return DETUNING[-1] + rates[-1] * ( t - sum( TIME ) )
    
    def hamiltonian_t(t, args): 
        """ evaluate the hamiltonian at time t. """
        H0 = args[0]
        H1 = args[1]
        x = det_pos( t )
        return H0 + x * H1

    H0 = qu.Qobj( [[0,MAG/2,0],[MAG/2,0,TUNNEL],[0,TUNNEL,0]] )
    H1 = qu.Qobj( [[1/2,0,0],[0,1/2,0],[0,0,-1/2]] )
  
    args = [H0, H1]
    
    c_ops = []

    output = qu.mesolve(hamiltonian_t, INIT, tlist, c_ops, operators, args, options = options)

    return tlist, dlist, real( output.expect[0] ), real( output.expect[1] ), real( output.expect[2] ), D_axis, T_axis, output.final_state
```
