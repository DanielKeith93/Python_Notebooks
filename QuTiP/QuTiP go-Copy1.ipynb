```python
import qutip as qu
import seaborn as sb
%pylab inline
```

## try a single spin

* define 2 x basis kets
* define 6 x operators x,y,z (pi,pi/2)
* define more operators hadamard etc

then draw some states on bloch plots, apply the gates...


```python
up = qu.basis(2,0)
dn = qu.basis(2,1)

iden = qu.qeye(2)

xpi = qu.sigmax()
ypi = qu.sigmay()
zpi = qu.sigmaz()

xp2 = qu.rx(pi*0.5)
yp2 = qu.ry(pi*0.5)
zp2 = qu.rz(pi*0.5)

had = qu.hadamard_transform(N=1)


#plot
#f = figure(figsize=(5,5),facecolor='white')
b = qu.Bloch()
b.figsize = [4,4]
#b.vector_color = sb.hls_palette(6, l=.4, s=.6)
b.vector_color = sb.color_palette()
b.zlabel = [r"$\left|\uparrow\right\rangle $",r"$\left|\downarrow\right\rangle$"]
b.xlabel = [r"$ X $",r""]
b.ylabel = [r"$ Y $",r""]
#b.zlabel = ["$|\uparrow$","$|\downarrow$"]
b.add_states([yp2*up,xp2*xpi*up,up,yp2*dn,xp2*xpi*dn,dn])
b.show()
```

## try a two electron hamiltonian

* zeeman term (maybe two different values)
* interaction term

```python
#some numbers
#h   = 4.1356675e-15   #Plancks constant (eV s)
#hb  = 6.5821e-16     #Reduced Plancks constant (eV s)
#g   = 1.9985               #Electron g-factor
#mu  = 5.7890e-05           #Bohr Magneton [ev/T]
#g_e = mu * g /h        #Electron gyromagnetic ratio
#g_n = 17.235e6         #Nuclear gyromagnetic ratio


#build 2e system sub-hamiltonians
Hze = qu.tensor([zpi,iden])+qu.tensor([iden,zpi]) #for equal B0
Hze1 = qu.tensor([zpi,iden]) #unequal fields
Hze2 = qu.tensor([iden,zpi])
Hex = qu.tensor([zpi,zpi])+qu.tensor([ypi,ypi])+qu.tensor([xpi,xpi]) #dot product

#H = ((gB+dgB)*Hze1 + (gB-dgB)*Hze2 + tc*Hex).unit()

#fn to project a density matrix onto a single spin's x/y/z (pass in sigmax etc)
def proj(op,n,size=2):
    out = op
    for i in range(0,n):
        out = qu.tensor([iden,out])
    for i in range(n+1,size):
        out = qu.tensor([out,iden])
    return out
#custom projectors for the S/T bloch sphere (i think they're correct now)
UDpro = qu.ket2dm(qu.tensor([up,dn]))-qu.ket2dm(qu.tensor([dn,up]))
STpro = qu.ket2dm(qu.tensor([up,dn])+qu.tensor([dn,up]))-qu.ket2dm(qu.tensor([up,dn])-qu.tensor([dn,up]))
STpro = STpro / 2.0 #Qobj.unit() gets the normalisation wrong - the projectors are not supposed to be untitary?
UiDpro = qu.ket2dm(qu.tensor([up,dn])+1j*qu.tensor([dn,up]))-qu.ket2dm(qu.tensor([up,dn])-1j*qu.tensor([dn,up]))
UiDpro = UiDpro / 2.0

#make a stack of projectors to send to the solver
projs1 = [proj(xpi,0),proj(ypi,0),proj(zpi,0),proj(xpi,1),proj(ypi,1),proj(zpi,1),UDpro,UiDpro,STpro]

#a nice fn to plot some bloch spheres
def plot_2spins(trajs):
    f = figure(figsize=(18,5))
    
    p1 = f.add_subplot(3,3,1)
    plot(out1.times,trajs[0])
    legend(['OneX','OneY','OneZ'],loc='best')
    
    p2 = f.add_subplot(3,3,4)
    plot(out1.times,trajs[1])
    legend(['TwoX','TwoY','TwoZ'],loc='best')
    
    p5 = f.add_subplot(3,3,7)
    p5.set_color_cycle(sb.color_palette()[::-1])
    plot(out1.times,trajs[2])
    legend(['ExchX','ExchY','ExchZ'],loc='best')
    
    p3 = f.add_subplot(1,3,2, projection='3d')
    b = qu.Bloch(fig=f,axes=p3)
    b.zlabel = [r"$\left|\uparrow\right\rangle $",r"$\left|\downarrow\right\rangle$"]
    b.xlabel = [r"$ X $",r""]
    b.ylabel = [r"$ Y $",r""]
    b.vector_color = sb.color_palette()
    b.add_states([yp2*up,xp2*xpi*up,up])
    b.point_color = sb.color_palette('dark')[3:]
    b.add_points(trajs[0].transpose(),'l')
    b.add_points(trajs[1].transpose(),'l')
    b.render(f,p3)
    
    p4 = f.add_subplot(1,3,3, projection='3d')
    b = qu.Bloch(fig=f,axes=p4)
    b.zlabel = [r"$\left|T\right\rangle $",r"$\left|S\right\rangle$"]
    b.xlabel = [r"$\left| \uparrow \downarrow \right\rangle$",r""]
    b.ylabel = [r"",r""]
    b.shpere_color = '#FFDDFF'
    b.vector_color = sb.color_palette()[::-1]
    b.add_states([yp2*up,xp2*xpi*up,up])
    b.point_color = sb.color_palette('dark')
    b.add_points(trajs[2].transpose(),'l')
    b.render(f,p4)
```

```python
#########EDIT THESE THINGS FIRST###########

#change coefficients to see different dynamics
tc = 0  #10e-6/(2*h)       #tunnel coupling
gB = 10 #2.0*28e9                #B-field default
dgB = 0 #hyperfine or something

#initail state and duration(adjust to see different things)
init = (qu.tensor(up+dn,up+2*dn)).unit()
t_vec = linspace(0,20,100)

H = ((gB+dgB)*Hze1 + (gB-dgB)*Hze2 + tc*Hex).unit()


#do the sim
out1 = qu.sesolve(H,init,t_vec,projs1)

#pull out individual spin trajectories and the STbasis also
traj1 = array(out1.expect).transpose()[:,0:3]
traj2 = array(out1.expect).transpose()[:,3:6]
traj3 = array(out1.expect).transpose()[:,6:9]
trajs = [traj1,traj2,traj3]

#draw it
plot_2spins(trajs)
```

```python
#########EDIT THESE THINGS FIRST###########

#change coefficients to see different dynamics
tc = 0  #10e-6/(2*h)       #tunnel coupling
gB0 = 1 #2.0*28e9                #B-field default
gB1 = 1   #2.0*28e9                #B-field default
w = 2.15
dgB = 0 #hyperfine or something

#initail state and duration(adjust to see different things)
init = (qu.tensor(up,up)).unit()
t_vec = linspace(0,20,100)

def H_osc1_coeff(t, args):
    return cos(w*t)

H_osc0 = (gB0)*(qu.tensor([zpi,iden]))
H_osc1 = (gB1)*(qu.tensor([xpi,iden]))
H_osc = [H_osc0,[H_osc1,H_osc1_coeff]]

#do the sim
out1 = qu.mesolve(H_osc,init,t_vec,_,projs1)

#pull out individual spin trajectories and the STbasis also
traj1 = array(out1.expect).transpose()[:,0:3]
traj2 = array(out1.expect).transpose()[:,3:6]
traj3 = array(out1.expect).transpose()[:,6:9]
trajs = [traj1,traj2,traj3]

#draw it
plot_2spins(trajs)
```
