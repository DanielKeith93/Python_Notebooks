```python
import qutip as qu
import seaborn as sb
```

```python
#basis states
up = qu.basis(2,0)
dn = qu.basis(2,1)

iden = qu.qeye(2)

#useful operators
sx = qu.sigmax()
sy = qu.sigmay()
sz = qu.sigmaz()
```

```python
#initialise state
init = (qu.tensor(up+2*dn,dn+2*up)).unit()
tlist = linspace(0,10,1500)
```

```python
#hamiltonian
H = qu.tensor(sz, iden) + qu.tensor(iden, sz) + 0.05 * qu.tensor(sx, sx)
H
```

```python
#collapse operators
gamma1 = 0.0 * 2 * pi           # relaxation rate
n_th = 0.0 * 2 * pi             # average number of thermal photons
c_ops = []

rate = gamma1 * (1 + n_th)
if rate > 0.0:
    c_ops.append(sqrt(rate) * sm)       # relaxation

rate = gamma1 * n_th
if rate > 0.0:
    c_ops.append(sqrt(rate) * sm.dag()) # excitation
```

```python
#solve with master equation
output = qu.mesolve(H, init, tlist, c_ops, [qu.tensor(sx,iden),qu.tensor(sy,iden),qu.tensor(sz,iden),qu.tensor(iden,sx),qu.tensor(iden,sy),qu.tensor(iden,sz)])
```

```python
traj1 = array(output.expect).transpose()[:,0:3]
traj2 = array(output.expect).transpose()[:,3:6]
```

```python
#plot trajectories
b = qu.Bloch()
b.figsize = [4,4]
b.zlabel = [r"$\left|\uparrow\right\rangle $",r"$\left|\downarrow\right\rangle$"]
b.xlabel = [r"$ X $",r""]
b.ylabel = [r"$ Y $",r""]
b.vector_color = sb.color_palette()
b.point_color = sb.color_palette('dark')[3:]
b.add_points(traj1.transpose(),'l')
b.add_points(traj2.transpose(),'l')
b.show()
```
