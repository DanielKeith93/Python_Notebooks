```python
import qutip as qu
import seaborn as sb
```

```python
init = ( 2 * qu.basis(2,0) + 1 * qu.basis(2,1)).unit()
init
```

```python
sx = qu.sigmax()
sy = qu.sigmay()
sz = qu.sigmaz()
```

```python
A = 0.2 * 2 * pi
tlist = linspace(0, 10, 1000)
H = A * sz
```

```python
gamma1 = 0.0 * 2 * pi           # relaxation rate
n_th = 0.0 * 2 * pi             # average number of thermal photons
c_ops = []

rate = gamma1 * (1 + n_th)
if rate > 0.0:
    c_ops.append(sqrt(rate) * sm)       # relaxation

rate = gamma1 * n_th
if rate > 0.0:
    c_ops.append(sqrt(rate) * sm.dag()) # excitation
```

```python
output = qu.mesolve(H, init, tlist, c_ops, [sx,sy,sz])
```

```python
traj = array(output.expect).transpose()[:,0:3]
```

```python
f = figure(figsize=(12,5))
p1 = f.add_subplot(121, projection='3d')
p2 = f.add_subplot(122)
b = qu.Bloch(fig=f,axes=p1)
b.zlabel = [r"$\left|\uparrow\right\rangle $",r"$\left|\downarrow\right\rangle$"]
b.xlabel = [r"$ X $",r""]
b.ylabel = [r"$ Y $",r""]
b.vector_color = sb.color_palette()
b.point_color = sb.color_palette('dark')[3:]
b.add_points(traj.transpose(),'l')
b.render(f,p1)
p2.plot( tlist, traj.transpose()[0], tlist, traj.transpose()[1], tlist, traj.transpose()[2] )
show()
```
