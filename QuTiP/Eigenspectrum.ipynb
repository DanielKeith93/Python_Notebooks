```python
import qutip as qu
import seaborn as sb
```

```python
#SInglet Hamiltonian 2 level

# set up the parameters
#
t_c = 0.1 * 2 * pi   # qubit sigma_x coefficient
Delta = 5.0 * 2 * pi   # detuning
time = 50   # sweep time
A = Delta / time   #sweep rate
gamma1 = 0.0005 * 2 * pi           # relaxation rate
n_th = 0.0 * 2 * pi             # average number of thermal photons
psi0 = (1)*(1*qu.basis(2,0)+0*qu.basis(2,1))      # initial state

H0 = - t_c/2.0 * sx
H1 = - A/2.0 * sz
args = (H0, H1)
#
# Hamiltonian
#
sx = qu.sigmax()
sy = qu.sigmay()
sz = qu.sigmaz()
sm = qu.destroy(2)

dlist = linspace(-Delta, Delta, 1500 )
tlist = linspace(-time, time, 1500)

eigvals = []
for i in tlist:
    H = H0 + H1 * i
    eigvals.append( qu.Qobj.eigenenergies(H) )
plot( dlist, eigvals )
xlabel('Detuning')
ylabel('Energy')
show()
```

```python
#Singlet Hamiltonian 3 level

X = 1.0 * 2 * pi
t = 5.0 * 2 * pi
V = 20.0 * 2 * pi

Delta = 10.0 * 2 * pi   # detuning
time = 10.0   # sweep time

A = Delta / time

H0 = qu.Qobj([[0, X, -sqrt(2)*t],[X, 0, -sqrt(2)*t ],[-sqrt(2)*t,-sqrt(2)*t,V]])
H1 = qu.Qobj([[-A,0,0],[0,A,0],[0,0,0]])

dlist = linspace(-Delta, Delta, 1500 )
tlist = linspace(-time, time, 1500)

eigvals = []
for i in tlist:
    H = H0 + H1 * i
    eigvals.append( qu.Qobj.eigenenergies(H) )
plot( dlist, eigvals )
xlabel('Detuning')
ylabel('Energy')
show()
```

```python
#Triplet Hamiltonian

bx = 1.0 * 2 * pi
bz = 1.0 * 2 * pi
V = 1.0 * 2 * pi

Delta = 10.0 * 2 * pi   # detuning
time = 10.0   # sweep time

A = Delta / time

H0 = qu.Qobj([[V+2*bz, bx*sqrt(2), 0],[bx*sqrt(2), V, bx*sqrt(2) ],[0,bx*sqrt(2)*t,V-2*bz]])

dlist = linspace(-Delta, Delta, 1500 )
tlist = linspace(-time, time, 1500)

eigvals = []
for i in tlist:
    H = H0
    eigvals.append( qu.Qobj.eigenenergies(H) )
plot( dlist, eigvals )
xlabel('Detuning')
ylabel('Energy')
show()
```
