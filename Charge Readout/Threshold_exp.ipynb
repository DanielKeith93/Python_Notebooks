```python
trace = []
time = []
signal = []
times = []
signals = []
traces = []

t1 = 0.1
t2 = 0.2
T = 0.35

#get traces from data file
import os
for file in os.listdir( "C:/Users/z3414503/Desktop/Honours/Data/144815_C Spin Tail Bz = 1.5000/traces" ):
    if file.endswith( ".dat" ):
        path = "C:/Users/z3414503/Desktop/Honours/Data/144815_C Spin Tail Bz = 1.5000/traces/"
        file = path + file
        d = qt.Data( file ) 
        x = d.get_data()
        trace, b, c = zip( *x )
        l = len( c ) / len( set( trace ) ) 
        l1 = t1 * l / T
        l2 = t2 * l / T
        for i in set( trace ): #split up the files into the seperate traces
            i = int( i )
            x = c[ ( i - 1 ) * l + int( l1 ) + 30: i * l + int( l1 ) + int( l2 ) - l - 30 ]
            y = b[ ( i - 1 ) * l + int( l1 ) + 30: i * l + int( l1 ) + int( l2 ) - l - 30 ] 
            signals.append( x )
            times.append( y )
            traces.append( trace[ i ] ) 
```

```python
#define function to generate the histograms with respect to the voltage threshold
def hist( S ):
    signals_down = []
    signals_up = []
    bars = 50
    V = V_t
    up = 0
    down = 0
    for x in S:
        if max( x ) >= V:
            signals_up.append( max( x ) )
            up += 1
        if max( x ) < V:
            signals_down.append( max( x ) )
            down += 1
    values_d, bins_d, _ = plt.hist( signals_down, bars, histtype = 'step', normed = True, color = 'r' )
    values_u, bins_u, _ = plt.hist( signals_up, bars, histtype = 'step', normed = True, color = 'b' )
    print up
    print down
    return values_d, bins_d, values_u, bins_u
```

```python
#plot the visibility for a range of current thresholds
V_t_max = 0.3 #Volts
V_t = linspace( 0, V_t_max, 100 )
Vis = []
F0 = []
F1 = []
for x in V_t:
    idx_d = min(range(len(bins_d)), key=lambda i: abs(bins_d[i]-x))
    idx_u = min(range(len(bins_u)), key=lambda i: abs(bins_u[i]-x))
    F_0 = 1 - sum( numpy.diff( bins_u[ :idx_u + 1 ] ) * values_u[ :idx_u ] )
    F_1 = 1 - sum( numpy.diff( bins_d[ idx_d: ] ) * values_d[ idx_d: ] )
    Vis.append( (F_0 + F_1) - 1 )
    F0.append( F_0 )
    F1.append( F_1 )
figure()
plot( V_t, Vis, V_t, F0, V_t, F1 )
ylabel( 'Visibility ($\%$)', fontsize = 20 )
xlabel( '$V_t (V)$', fontsize = 20 )
Vis_max = max( Vis )
print Vis_max * 100
idx = Vis.index( Vis_max )
print V_t[idx]
```
