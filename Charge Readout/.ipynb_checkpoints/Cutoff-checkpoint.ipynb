```python
#define parameters
T = 0.018
N_0 = 0.37
N_1 = 1 - N_0
deltat = T #s, taken from spin to charge optimisation
t_down_in = 0.0069
t_up_out = 0.0039 #Taken from experimental data.
deltaI = 32 * 10 ** - 12 #A
a_n = 71 * 10 ** - 15 #A / ( ( Hz ) ^ ( 1 / 2 ) ) #Have to multiply spectral density by a bandwidth.
bars = 200
traces = 1000
```

```python
f_c = logspace( -1, 5, 100)
```

```python
from scipy.signal import bessel, lfilter

def bessel_lowpass(lowcut, sample_rate, order=8):
    nyq = 0.5 * sample_rate
    low = lowcut / nyq
    b, a = bessel(order, low, btype='low')
    return b, a

def bessel_lowpass_filter(data, lowcut, sample_rate, order=8):
    b, a = bessel_lowpass(lowcut, sample_rate, order=order)
    y = lfilter(b, a, data)
    return y

def sqr_pulse( c, time_arr, t1, t2, I ):
    start = int( ( t1 / deltat ) * len(time_arr) )
    finish = int( ( ( t1 + t2 ) / deltat ) * len(time_arr) )
    if c == 0:
        wave = time_arr[ : ] * 0
        wave[ start: finish ] =  ( time_arr[ start: finish ] * 0 ) + I 
        return wave
    elif c == 1:
        wave = time_arr[ : ] * 0
        return wave
    
```

```python
def gen_wav3( N, freq, sample_rate, noise, no_points ):
    N_down = []
    N_up = [] #up goes to charge 0
    up = 0
    down = 0
    for i in xrange( 0, N):
        if randint(0,99) < ( N_0 * 100 ):
            charge = 0
        else:
            charge = 1
            #need to randomly generate different tunneling times for each simulated pulse.
        t_in = t_down_in 
        t_out = t_up_out
        r_in = randint( 1,1001 )
        r_out = randint( 1,1001 )
        t_in = - log( r_in * 0.001 ) * t_in
        t_out = - log( r_out * 0.001 ) * t_out
        if charge == 0:
            N_up.append( max( bessel_lowpass_filter(sqr_pulse( charge, time1, t_out, t_in, deltaI ) + noise * random.normal( 0, 1, no_points ), freq , sample_rate, order=8)) )
            up += 1
        else:
            N_down.append( max( bessel_lowpass_filter(sqr_pulse( charge, time1, t_out, t_in, deltaI ) + noise * random.normal( 0, 1, no_points ), freq , sample_rate, order=8)) )
            down += 1
    hist( N_down, bars, histtype = 'step', normed = True )
    hist( N_up, bars, histtype = 'step', normed = True )
    values_d, bins_d, _ = hist(N_down, bars, histtype = 'step', normed = True)
    values_u, bins_u, _ = hist(N_up, bars, histtype = 'step', normed = True)

    #print up
    #print down

    return values_d, bins_d, values_u, bins_u
```

```python
time1 = linspace( 0, T, 600 * T )
y = sqr_pulse( 0, time1, t_up_out, t_down_in, deltaI )
x = a_n * sqrt( 600 ) * random.normal( 0, 1, 600 * T )
z = random.normal( 0, 1, 600 * T )
figure()
plot( time1, y ,'r')
figure()
plot( time1, x, 'g' )
figure()
plot( time1, z, 'b' )
figure()
values_d, bins_d, values_u, bins_u = gen_wav3( 1000, 300, 600, a_n * sqrt( 600 ) , 600 * T )
```

```python
from scipy.signal import bessel, lfilter

f_c = logspace( 2, 5, 100)
Vmax = []
for i in f_c: 
    #print i
    fs = 2 * i
    A_n = a_n * sqrt( fs )
    points = T * fs
    time1 = linspace( 0, deltat, points )

    #print i
    #print A_n
    #print traces
    #print fs
    #print points
    values_d, bins_d, values_u, bins_u = gen_wav3( traces, i, fs, A_n, points )

    I_t_max = 6 * 10 ** - 10
    I_t = linspace( 0, I_t_max, 100 )
    V = []
    for x in I_t:
        idx_d = min(range(len(bins_d)), key=lambda i: abs(bins_d[i]-x))
        idx_u = min(range(len(bins_u)), key=lambda i: abs(bins_u[i]-x))
        F_0 = 1 - sum( numpy.diff( bins_u[ :idx_u + 1 ] ) * values_u[ :idx_u ] )
        F_1 = 1 - sum( numpy.diff( bins_d[ idx_d: ] ) * values_d[ idx_d: ] )
        V.append( (F_0 + F_1) - 1 )
    V_max = max( V )
    #print V_max * 100
    Vmax.append( V_max * 100 )
figure()
plot( f_c, Vmax )
xscale( 'log' )
```

```python
Vmax = []
I_tmax = []
f_c = logspace( 1.5, 6, 500)
I_t_max = 6 * 10 ** - 10
I_t = linspace( 0, I_t_max, 250 )
for i in f_c:
    fs = 2 * i
    A_n = a_n * sqrt( fs )
    points = T * fs
    time1 = linspace( 0, deltat, points )
    values_d, bins_d, values_u, bins_u = gen_wav3( traces, i, fs, A_n, points )
    V = []
    for x in I_t:
        idx_d = min(range(len(bins_d)), key=lambda i: abs(bins_d[i]-x))
        idx_u = min(range(len(bins_u)), key=lambda i: abs(bins_u[i]-x))
        F_0 = 1 - sum( numpy.diff( bins_u[ :idx_u + 1 ] ) * values_u[ :idx_u ] )
        F_1 = 1 - sum( numpy.diff( bins_d[ idx_d: ] ) * values_d[ idx_d: ] )
        V.append( (F_0 + F_1) - 1 )
    #figure()
    #plot( I_t, V )
    V_max = max( V )
    Vmax.append( V_max )
    #print V_max * 100
    idx = V.index( V_max )
    I_tmax.append( I_t[ idx ] )
    #print I_t[idx]
print( max( Vmax ) )
figure()
plot( f_c, Vmax )
plt.axhline(y=.95, c = 'r', linestyle = '--' )
idx = ( abs( Vmax - max( Vmax ) ) ).argmin()
plt.axvline(x= f_c[ idx ], c = 'k' )
xlabel( 'cutoff frequency (Hz)', fontsize=20 )
ylabel( 'Visibility (fraction)', fontsize=20 )
xscale( 'log' )
figure()
plot( f_c, I_tmax )
xlabel( 'cutoff frequency (Hz)', fontsize=20 )
ylabel( 'current threshold (A)', fontsize=20 )
xscale( 'log' )
```
