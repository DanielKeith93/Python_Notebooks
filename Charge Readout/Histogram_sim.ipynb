```python
fs = 20000.0
lowcut = 10000.0
T = 0.055
nsamples = T * fs
from scipy.signal import bessel, lfilter


def bessel_lowpass(lowcut,fs, order=8):
    nyq = 0.5 * fs
    low = lowcut / nyq
    b, a = bessel(order, low, btype='low')
    return b, a


def bessel_lowpass_filter(data, lowcut,fs, order=8):
    b, a = bessel_lowpass(lowcut,fs, order=order)
    y = lfilter(b, a, data)
    return y
```

```python
#create filter function
import scipy.signal as sig

#a function to filter a trace
def lowpass(times,data,cutoff):
    samprate = 1.0/(time[1]-time[0])
    b,a = sig.bessel(8,cutoff/(samprate/2.0),btype='low',analog=0,output='ba')
    #note cutoff is given relative to nyquist(=sample rate/2)
    data_f = sig.filtfilt(b,a,data)
    return data_f
```

```python
#define parameters
N_0 = 0.55
N_1 = 1 - N_0
deltat = 0.055 #s, taken from spin to charge optimisation
t_down_in = 0.0051
t_up_out = 0.0065 #Taken from experimental data.
deltaI = 0.45 * 10 ** - 9 #A
points = nsamples
time = linspace( 0, deltat, points )
a_n = 0.24 * 10 ** - 12 #A / ( ( Hz ) ^ ( 1 / 2 ) ) #Have to multiply spectral density by a bandwidth.
f_c = 10000 #Hz, cut off frequency, also used as the bandwidth
A_n = a_n * sqrt( fs )
```

```python
# a function to make square pulses
def sqr_pulse( c, time_arr, t1, t2, I ):
    '''sqr_pulse(charge, time_arr, t1, t2, I)
    IN: c = int, 0 or 1
        time_arr = array-like of floats defining time-points
        t1 = delay (pulse switches on)
        t2 = duration (between on and off)
        I = peak current
    OUT: array-like pulse matching the size of time_arr
    '''
    
    start = int( ( t1 / deltat ) * len(time_arr) )
    finish = int( ( ( t1 + t2 ) / deltat ) * len(time_arr) )
    if c == 0:
        wave = time_arr[ : ] * 0
        wave[ start: finish ] =  ( time_arr[ start: finish ] * 0 ) + I 
        return wave
    elif c == 1:
        wave = time_arr[ : ] * 0
        return wave
        
```

```python
#generate random waveforms and measures the peak currents I_p
def gen_wav2( N ):
    N_down = []
    N_up = [] #up goes to charge 0    
    for i in xrange( 0, N):
        spin = randint(0,99)
        if spin < ( N_0 * 100 ):
            charge = 0
        else:
            charge = 1
        wave = sqr_pulse( charge, time, t_up_out, t_down_in, deltaI )
        noise = random.normal( 0, A_n, points )
        noisy_wave = wave + noise
        filtered_wave2 = bessel_lowpass_filter(noisy_wave, lowcut,fs, order=8)
        I_p = max(filtered_wave2)
        if charge == 0:
            N_up.append( I_p )
        else:
            N_down.append( I_p )
    hist( N_down, 200, histtype = 'step', normed = True )
    hist( N_up, 200, histtype = 'step', normed = True )
    values, bins, _ = plt.hist(N_down, 200, histtype = 'step', normed = True)

    return values, bins
```

```python
#generate random waveforms and measures the peak currents I_p
#compacted the code and made to return values
bars = 200
def gen_wav2( N ):
    N_down = []
    N_up = [] #up goes to charge 0    
    for i in xrange( 0, N):
        if randint(0,99) < ( N_0 * 100 ):
            charge = 0
        else:
            charge = 1
            #need to randomly generate different tunneling times for each simulated pulse.
        if charge == 0:
            N_up.append( max( bessel_lowpass_filter(sqr_pulse( charge, time, t_up_out, t_down_in, deltaI ) + random.normal( 0, A_n, points ), lowcut,fs, order=8)) )
        else:
            N_down.append( max( bessel_lowpass_filter(sqr_pulse( charge, time, t_up_out, t_down_in, deltaI ) + random.normal( 0, A_n, points ), lowcut,fs, order=8)) )
    hist( N_down, bars, histtype = 'step', normed = True )
    hist( N_up, bars, histtype = 'step', normed = True )
    values_d, bins_d, _ = plt.hist(N_down, 200, histtype = 'step', normed = True)
    values_u, bins_u, _ = plt.hist(N_up, 200, histtype = 'step', normed = True)


    return values_d, bins_d, values_u, bins_u
```
