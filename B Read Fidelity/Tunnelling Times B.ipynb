```python
number_of_files = 20

V = 0.032

start = 0.0262
finish = 0.5246

up_outs = []
down_ins = []
up_ins = []
down_outs = []

trace = []
traces = []

time = linspace( 0.0, 0.4984, 2490)

for x in xrange( 1, number_of_files + 1 ):
    print x,
    file = "C:/Users/z3414503/Desktop/Honours/Data/B Read Data/TRACES/Read B at Bz = 2.5000_" + str(x) + "_trace.dat"
    with open( file, "r" ) as f:
        i = 1
        for line in f:
            fields = line.split()
            if len(fields) == 3 and fields[0] != '#':
                if int( fields[0] ) == i:
                    value = 0.0
                    if float( fields[1] ) >= start and float( fields[1] ) <= finish and float( fields[1] ) >= value and len( trace ) < 2490:
                        trace.append( float( fields[2] ) )
                        value = float( fields[1] )
                if int( fields[0] ) > i:
                    i += 1
                    traces.append(trace)
                    for l in trace:
                        if l > V:
                            idx1=trace.index(l)
                            up_outs.append( time[idx1] )
                            for j in trace[ idx1: ]:
                                if j < 0:
                                    idx2=trace[ idx1: ].index( j )
                                    down_ins.append( time[idx2] )
                                    for k in trace[ idx1 + idx2: ]:
                                        if k > V:
                                            up_ins.append( down_ins[-1] )
                                            del down_ins[-1]
                                            break
                                    break
                            break
                    trace = []
    trace = []
```
