```python
import scipy
from scipy import stats
from __future__ import division
import scipy.fftpack
from scipy.stats import norm
from scipy.fftpack import rfft, irfft, fftfreq
```

```python
#constants used
g_e = 2
mu_B = 5.7884 * 10 ** - 5 
k_B = 8.6173 * 10 ** - 5 
bars = 200
no_traces = 1000
#set variables
t_up_out = 0.0316
t_down_in = 0.0077
B = 2.5
N_up_0 = 0.41
f_c = 1000.0
fs = 5000
deltaT = 0.209
deltaI = 0.050 #V
a_n = 1.4849 * 10 ** - 4 #V / ( ( Hz ) ^ ( 1 / 2 ) ) #Have to multiply spectral density by a bandwidth.
#calculate other parameters
N_down_0 = 1 - N_up_0
t_down_out = 25
T1 = 8.4
points = deltaT * fs
T2 = T1 * t_down_out - T1 * t_up_out + t_down_out * t_up_out
```

```python
from scipy.signal import bessel, lfilter
from numba import jit

@jit
def filter( TIME, SIGNAL, F ):
    W = fftfreq(size( SIGNAL ), d=TIME[1]-TIME[0])
    f_signal = rfft(SIGNAL)
    
#     print size(SIGNAL)
#     print len(W)
#     print len(f_signal)

    # If our original signal time was in seconds, this is now in Hz    
    cut_f_signal1 = f_signal.copy()
    cut_f_signal1[(W<F)] = 0

    cut_signal1 = irfft(cut_f_signal1)
    
    return SIGNAL - cut_signal1

@jit
def sqr_pulse( c, time_arr, t1, t2, I ):
    start = int( ( t1 / deltaT ) * len(time_arr) )
    finish = int( ( ( t1 + t2 ) / deltaT ) * len(time_arr) )
    if c == 0:
        wave = time_arr[ : ] * 0
        wave[ start: finish ] =  ( time_arr[ start: finish ] * 0 ) + I
        return wave
    elif c == 1:
        wave = time_arr[ : ] * 0
        return wave
@jit    
def gen_wav3( N, noise, fraction, height, cutoff ):
    N_down = []
    N_up = [] #up goes to charge 0
    for i in xrange( 0, N):
        if randint(0,99) < ( fraction * 100 ):
            charge = 0
        else:
            charge = 1
            #need to randomly generate different tunneling times for each simulated pulse.
        t_in = t_down_in 
        t_out = t_up_out
        r_in = randint( 1,1001 )
        r_out = randint( 1,1001 )
        t_in = - log( r_in * 0.001 ) * t_in
        t_out = - log( r_out * 0.001 ) * t_out
        if charge == 0:
            N_up.append( max( filter(time1,sqr_pulse( charge, time1, t_out, t_in, height ) + noise * random.normal( 0, 1, points ), cutoff ) ) )
        else:
            N_down.append( max( filter(time1,sqr_pulse( charge, time1, t_out, t_in, height ) + noise * random.normal( 0, 1, points ), cutoff ) ) )

    range = deltaI + 5 * noise
    
    values_u, bins_u = histogram( N_up, bins=arange( -0.1, 0.2, 0.001 ), normed = True )
    values_d, bins_d = histogram( N_down, bins=arange( -0.1, 0.2, 0.001 ), normed = True )

    return values_d, bins_d, values_u, bins_u
```

```python

f_c = logspace( 0, 5, 500)
I_t_max = 0.1
I_t = linspace( 0.02, I_t_max, 1000 )

sample_rates = [2500,5000,7500,10000]
visibilities = []

# for j in sample_rates:
#     fs = j
Vmax = []
I_tmax = []
for i in f_c:
    fs = 2 * i

    A_n = a_n * sqrt( fs )
#     points = T * fs
    time1 = linspace( 0, deltaT, points )
    values_d, bins_d, values_u, bins_u = gen_wav3( no_traces, A_n, N_up_0, deltaI, i )
    V = []
    for x in I_t:
        idx_d = min(range(len(bins_d)), key=lambda i: abs(bins_d[i]-x))
        idx_u = min(range(len(bins_u)), key=lambda i: abs(bins_u[i]-x))
        F_0 = 1 - sum( numpy.diff( bins_u[ :idx_u + 1 ] ) * values_u[ :idx_u ] )
        F_1 = 1 - sum( numpy.diff( bins_d[ idx_d: ] ) * values_d[ idx_d: ] )
        V.append( (F_0 + F_1) - 1 )
    #figure()
    #plot( I_t, V )
    V_max = max( V )
    Vmax.append( V_max )
    #print V_max * 100
    idx = V.index( V_max )
    I_tmax.append( I_t[ idx ] )
    #print I_t[idx]
# print( max( Vmax ) )
# visibilities.append( Vmax )
```
