```python
from scipy import signal

#define parameters

fs = 10000.0 #Hz sample rate
lowcut = 1000.0 #Hz filter cutoff frequency
T = 0.0118 #s readout time
nsamples = T * fs #number of points in trace
N_0 = 0.45 #spin up fraction
N_1 = 1 - N_0
deltat = T #s, taken from spin to charge optimisation
t_down_in = 0.0052 #characteristic tunnelling times
t_up_out = 0.00107 
deltaI = 0.023 #V signal height
points = nsamples
time1 = linspace( 0, deltat, points ) #time axis for traces
a_n = 0.24 * 10 ** - 12 #V / ( ( Hz ) ^ ( 1 / 2 ) ) #Have to multiply spectral density by a bandwidth.
f_c = lowcut #Hz, cut off frequency, also used as the bandwidth
A_n = 0.0076 #V a_n * sqrt( fs )
bars = 200 #bars in histograms
traces = 5000 #number of traces to simulate
```

```python
from scipy.signal import bessel, lfilter

#simulate the frequency filter used in actual experiments. Functions don't really work and should be replaced
#by better filter function from other notebooks

def bessel_lowpass(lowcut,fs, order=8):
    nyq = 0.5 * fs
    low = lowcut / nyq
    b, a = bessel(order, low, btype='low')
    return b, a


def bessel_lowpass_filter(data, lowcut,fs, order=8):
    b, a = bessel_lowpass(lowcut,fs, order=order)
    y = lfilter(b, a, data)
    return y

# a function to make square pulses of specified dimensions

def sqr_pulse( c, time_arr, t1, t2, I ):
    start = int( ( t1 / deltat ) * len(time_arr) )
    finish = int( ( ( t1 + t2 ) / deltat ) * len(time_arr) )
    if c == 0:
        wave = time_arr[ : ] * 0
        wave[ start: finish ] =  ( time_arr[ start: finish ] * 0 ) + I 
        return wave
    elif c == 1:
        wave = time_arr[ : ] * 0
        return wave

#wavelet used to perform the continuous wavelet transform
    
def DOG(a, length):
    t = np.linspace(-length/2, (length)/2, length)
    y = - a ** 2 * t * exp( - t ** 2 / ( 2 * a ** 2 ) ) #/ ( sqrt( 2 * pi ) * a ** 3 )
    return y

#smooths out unecessary peaks in a trace that slow down calculations

def smooth(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth
```

```python
#generate random waveforms and measures the peak currents I_p
#compacted the code and made to return values
def gen_wav3( N, threshold ):
    N_down = []
    N_up = [] #up goes to charge 0
    charge = 0
    up = 0
    for x in xrange( 0, N ):
        t_in = t_down_in 
        t_out = t_up_out
        r_in = randint( 1,1001 )
        r_out = randint( 1,1001 )
        t_in = - log( r_in * 0.001 ) * t_in #random tunnelling times to determine dimensions
        t_out = - log( r_out * 0.001 ) * t_out
        
        #create simulated trace
    
        sig = bessel_lowpass_filter(sqr_pulse( charge, time1, t_out, t_in, deltaI ) + A_n * random.normal( 0, 1, points ), lowcut,fs, order=8)
        if edge_det( sig, threshold ) == 0: #determine if the maxmimum weight is above an arbitrary threshold
            N_up.append( max( sig ) )
            up += 1
    
    return up
```

```python
#determines if a tunnelling event is present in a trace based on the continuous wavelet transform

def edge_det( z, threshold ):
    sig  = z
    widths = np.arange(1, 101)
    cwtmatr = smooth( signal.cwt(sig, DOG, widths), 9 )
    peaks = []
    for i in cwtmatr: 
        max_min = []
        b = i
        for x in xrange( 1, 1099 ):
            if b[x-1]>b[x] and b[x+1]>b[x]:
                max_min.append( x )
            if b[x-1]<b[x] and b[x+1]<b[x]:
                max_min.append( x )
        peaks.append( max_min )
    tracks = []
    for i in peaks[ -1 ]:
        track = []
        x = i
        track.append( x )
        for s in xrange( 1, 100 ):
            a = peaks[ - s - 1 ]
            idx = min(range(len(a)), key=lambda i: abs(a[i]-x))
            x = peaks[ - s - 1 ][ idx ]
            track.append( x )
        tracks.append( track )
    norm = []
    y = []
    for l in xrange( 0, 100 ):
        for k in cwtmatr[ l ]:
            a = k ** 2
            y.append( a )
        med = median( y )
        norm.append( med )
    weights = []
    for i in xrange( 0, len( peaks[ -1 ] ) ):
        weight = []
        for j in tracks[ i ]:
            scale = 99
            a = cwtmatr[ scale ][ j ] ** 2
            b = norm[ scale ]
            weight.append( a / b )
            scale -= 1
        c = sum( weight )    
        weights.append( c )
    #plot( peaks[ -1 ], weights, 'o' )
    steps = []
    for index,value in enumerate(weights):
        if value > threshold:
            i = index
            steps.append( peaks[-1][i] )
    steps2 = []
    for i in steps:
        a = cwtmatr[ 99 ][ i ]
        steps2.append( a )
    steps3 = []
    for i in xrange(0, len(steps2)):
        if steps2[ i ] == max( steps2 ):
            steps3.append( steps[ i ] )
        if steps2[ i ] == min( steps2 ):
            steps3.append( steps[ i ] )
    if len( steps3 ) == 0:
        return 1
    if len( steps3 ) == 2:
        return 0
```

```python
#determines the maximum weight of a trace using the continuous wavelet transform

def edge_max( z ):
    sig  = z
    widths = np.arange(1, 101)
    cwtmatr = smooth( signal.cwt(sig, DOG, widths), 9 )
    peaks = []
    for i in cwtmatr: 
        max_min = []
        b = i
        for x in xrange( 1, 1099 ):
            if b[x-1]>b[x] and b[x+1]>b[x]:
                max_min.append( x )
            if b[x-1]<b[x] and b[x+1]<b[x]:
                max_min.append( x )
        peaks.append( max_min )
    tracks = []
    for i in peaks[ -1 ]:
        track = []
        x = i
        track.append( x )
        for s in xrange( 1, 100 ):
            a = peaks[ - s - 1 ]
            idx = min(range(len(a)), key=lambda i: abs(a[i]-x))
            x = peaks[ - s - 1 ][ idx ]
            track.append( x )
        tracks.append( track )
    norm = []
    y = []
    for l in xrange( 0, 100 ):
        for k in cwtmatr[ l ]:
            a = k ** 2
            y.append( a )
        med = median( y )
        norm.append( med )
    weights = []
    for i in xrange( 0, len( peaks[ -1 ] ) ):
        weight = []
        for j in tracks[ i ]:
            scale = 99
            a = cwtmatr[ scale ][ j ] ** 2
            b = norm[ scale ]
            weight.append( a / b )
            scale -= 1
        c = sum( weight )    
        weights.append( c )
    return max( weights )
```

```python
#similar to gen_wav3 except that it uses edge_max() instead of edge_det()

def gen_wav4( n ):
    N_up = []
    N_down = []
    for i in xrange( 0, n):
        if randint(0,99) < ( N_0 * 100 ):
            charge = 0
        else:
            charge = 1
        t_in = t_down_in 
        t_out = t_up_out
        r_in = randint( 1,1001 )
        r_out = randint( 1,1001 )
        t_in = - log( r_in * 0.001 ) * t_in
        t_out = - log( r_out * 0.001 ) * t_out
        sig = bessel_lowpass_filter(sqr_pulse( charge, time1, t_out, t_in, deltaI ) + A_n * random.normal( 0, 1, points ), lowcut,fs, order=8)
        a = edge_max( sig )
        if charge == 0:
            N_up.append( a )
        else:
            N_down.append( a )

    return N_up, N_down
```

```python
#calculate the maximum weights of the up and down traces produced by the simulation

x, y = gen_wav4( 1000 )
```

```python
#calculate the fidelities for various different thresholds

w_t = logspace( 5, 9, 100000 ) #list of weight thresholds to test
V = []
F0 = []
F1 = []
for x in w_t:
    idx_d = min(range(len(bins_d)), key=lambda i: abs(bins_d[i]-x))
    idx_u = min(range(len(bins_u)), key=lambda i: abs(bins_u[i]-x))
    F_0 = 1 - sum( numpy.diff( bins_u[ :idx_u + 1 ] ) * values_u[ :idx_u ] ) #based on the area underneath the histograms to either side of the threshold
    F_1 = 1 - sum( numpy.diff( bins_d[ idx_d: ] ) * values_d[ idx_d: ] )
    V.append( (F_0 + F_1) - 1 )
    F0.append( F_0 )
    F1.append( F_1 )
figure()
xscale( 'log' )
plot( w_t, V, w_t, F0, w_t, F1 )
V_max = max( V ) #find at what threshold the visibility is a maximum
print V_max * 100
idx = V.index( V_max )
print w_t[idx]
```
