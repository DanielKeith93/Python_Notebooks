```python
@jit
def filter( TIME, SIGNAL, F ):
    W = fftfreq(size( SIGNAL ), d=TIME[1]-TIME[0])
    f_signal = rfft(SIGNAL)
    
    print size(SIGNAL)
    print len(W)
    print len(f_signal)

    # If our original signal time was in seconds, this is now in Hz    
    cut_f_signal1 = f_signal.copy()
    cut_f_signal1[(W<F)] = 0

    cut_signal1 = irfft(cut_f_signal1)
    
    return SIGNAL - cut_signal1
```

```python
traces_filt = []
for trace in traces:
    signal = trace

    W = fftfreq(size( signal ), d=time1[1]-time1[0])
    f_signal = rfft(signal)

    # If our original signal time was in seconds, this is now in Hz    
    cut_f_signal1 = f_signal.copy()
    cut_f_signal1[(W>305)] = 0
    cut_f_signal1[(W<295)] = 0

    cut_f_signal2 = f_signal.copy()
    cut_f_signal2[(W>905)] = 0
    cut_f_signal2[(W<895)] = 0

    cut_signal1 = irfft(cut_f_signal1)
    cut_signal2 = irfft(cut_f_signal2)
    
    traces_filt.append( signal - cut_signal1 - cut_signal2 )
```
