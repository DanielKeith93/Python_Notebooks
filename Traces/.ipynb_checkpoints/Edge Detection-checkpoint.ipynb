```python
from scipy import signal
#define parameters
fs = 20000.0 #Hz ample rate
lowcut = 10000.0 #Hz cutoff filter frequency
T = 0.055 #s readout time
nsamples = T * fs #number of points in trace
N_0 = 0.55 #spin up fraction
N_1 = 1 - N_0
deltat = T #s, taken from spin to charge optimisation
t_down_in = 0.0051 #characteristic tunnelling times
t_up_out = 0.0065 #Taken from experimental data.
deltaI = 0.45 * 10 ** - 9 #A signal height
points = nsamples
time1 = linspace( 0, deltat, points ) #time axis for traces
a_n = .24 * 10 ** - 12 #A / ( ( Hz ) ^ ( 1 / 2 ) ) #Have to multiply spectral density by a bandwidth.
f_c = lowcut #Hz, cut off frequency, also used as the bandwidth
A_n = a_n * sqrt( fs ) #noise dependent on sample rate
bars = 200 #to be used in histogram
traces = 10000 #number of traces created in simulation
```

```python
from scipy.signal import bessel, lfilter

#define functions to be used for filter, but this filter does not work. Needs to be replaced by updated filter function

def bessel_lowpass(lowcut,fs, order=8):
    nyq = 0.5 * fs
    low = lowcut / nyq
    b, a = bessel(order, low, btype='low')
    return b, a


def bessel_lowpass_filter(data, lowcut,fs, order=8):
    b, a = bessel_lowpass(lowcut,fs, order=order)
    y = lfilter(b, a, data)
    return y

#function produces the ideal square shape of the simulated traces of specified dimensions

def sqr_pulse( c, time_arr, t1, t2, I ):
    start = int( ( t1 / deltat ) * len(time_arr) )
    finish = int( ( ( t1 + t2 ) / deltat ) * len(time_arr) )
    if c == 0:
        wave = time_arr[ : ] * 0
        wave[ start: finish ] =  ( time_arr[ start: finish ] * 0 ) + I 
        return wave
    elif c == 1:
        wave = time_arr[ : ] * 0
        return wave
    
#derivative of a gaussian used as wavelet in the continuous wavelet transform, picks out large steps in the traces
   
def DOG(a, length):
    t = np.linspace(-length/2, (length)/2, length)
    y = - a ** 2 * t * exp( - t ** 2 / ( 2 * a ** 2 ) ) #/ ( sqrt( 2 * pi ) * a ** 3 )
    return y

#smoothes out traces to get rid of unimportant trace peaks that slow down calculations

def smooth(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth
```

```python
#check to see if simulated traces are being created correctly

#generated with random times to define the trace dimensions

r_in = randint( 1,1001 )
r_out = randint( 1,1001 )
t_in = - log( r_in * 0.001 ) * t_down_in
t_out = - log( r_out * 0.001 ) * t_up_out
z = bessel_lowpass_filter(sqr_pulse( 0, time1, t_out, t_in, deltaI ) + A_n * random.normal( 0, 1, points ), lowcut,fs, order=8)
z2 = sqr_pulse( 0, time1, t_out, t_in, deltaI )
plot( z )
plot( z2, 'r' )
```

```python
#can time how long a function takes to compute

import time
start = time.clock()

sig  = z
widths = np.arange(1, 101)
cwtmatr = signal.cwt(sig, DOG, widths) #take the transform of the desired trace
peaks = []
for i in cwtmatr: 
    max_min = []
    b = i
    for x in xrange( 1, 1099 ):
        if b[x-1]>b[x] and b[x+1]>b[x]:
            max_min.append( x )
        if b[x-1]<b[x] and b[x+1]<b[x]:
            max_min.append( x )
    peaks.append( max_min ) #create a list of all the peaks that occur in the transform
track = []
x = peaks[ -1 ][ 0 ] #a track will be found for each peak of the transform at the largest scale
track.append( x )
for s in xrange( 1, 100 ):
    a = peaks[ - s - 1 ]
    idx = min(range(len(a)), key=lambda i: abs(a[i]-x)) #find the closest peak to the peak of the previous scale
    x = peaks[ - s - 1 ][ idx ]
    track.append( x )
tracks = []
for i in peaks[ -1 ]: #same process for the other peaks
    track = []
    x = i
    track.append( x )
    for s in xrange( 1, 100 ):
        a = peaks[ - s - 1 ]
        idx = min(range(len(a)), key=lambda i: abs(a[i]-x))
        x = peaks[ - s - 1 ][ idx ]
        track.append( x )
    tracks.append( track )
norm = []
y = []
for l in xrange( 0, 100 ): #calculate the normalisation constant for each scale of the transform
    for k in cwtmatr[ l ]:
        a = k ** 2
        y.append( a )
    med = median( y )
    norm.append( med )
weights = []
for i in xrange( 0, len( peaks[ -1 ] ) ):
    weight = []
    for j in tracks[ i ]:
        scale = 99
        a = cwtmatr[ scale ][ j ] ** 2
        b = norm[ scale ]
        weight.append( a / b )
        scale -= 1
    c = sum( weight )    #sum up all the normalised values of the transform along each track to produce the weight of each peak
    weights.append( c )
    
print time.clock() - start
```

```python
#find all the weights of a trace that are above an arbitrary threshold and calculate the value of the transform at each point
#positive value of the transform correspond to a step up and negative corresponds to a step down.

steps = []
for index,value in enumerate(weights):
    if value > 40000000:
        i = index
        steps.append( peaks[-1][i] )
print steps
steps2 = []
for i in steps:
    a = cwtmatr[ 99 ][ i ]
    steps2.append( a )
print steps2
steps3 = []
for i in xrange(0, len(steps2)):
    if steps2[ i ] == max( steps2 ):
        steps3.append( steps[ i ] )
    if steps2[ i ] == min( steps2 ):
        steps3.append( steps[ i ] )
print steps3
```
