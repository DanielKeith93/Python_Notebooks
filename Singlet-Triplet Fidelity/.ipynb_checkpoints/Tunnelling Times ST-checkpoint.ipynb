```python
import sys
sys.path[0] = 'C:/Users/z3414503/Documents/QTLab/qtlab_013/source/' 
import qt
import scipy
from scipy import stats
from __future__ import division
```

```python
number_of_files = 20

V = 0.005

start = 0.001
finish = 0.5

up_outs = []
down_ins = []
up_ins = []
down_outs = []

trace = []
traces = []

time = linspace( 0.0, 0.449, 4990)

for x in xrange( 1, number_of_files + 1 ):
    print x,
    file = "C:/Users/z3414503/Desktop/Honours/Data/ST Read/TRACES/Read ST at Bz = 2.5000_" + str(x) + "_trace.dat"
    with open( file, "r" ) as f:
        i = 1
        for line in f:
            fields = line.split()
            if len(fields) == 3 and fields[0] != '#':
                if int( fields[0] ) == i:
                    value = 0.0
                    if float( fields[1] ) >= start and float( fields[1] ) < finish and float( fields[1] ) >= value and len( trace ) < 4991:
                        trace.append( float( fields[2] ) )
                        value = float( fields[1] )
                if int( fields[0] ) > i:
                    i += 1
                    traces.append(trace)
                    for l in trace:
                        if l < V:
                            idx1=trace.index(l)
                            up_outs.append( time[idx1] )
                            for j in trace[ idx1: ]:
                                if j > 0.015:
                                    idx2=trace[ idx1: ].index( j )
                                    down_ins.append( time[idx2] )
                                    for k in trace[ idx1 + idx2: ]:
                                        if k < V:
                                            up_ins.append( down_ins[-1] )
                                            del down_ins[-1]
                                            break
                                    break
                            break
                    trace = []
    trace = []
```

```python
U_O = []
for i in up_outs:
    if i < 0.05 and i > 0.001:
        U_O.append( i )
```

```python
param = stats.expon.fit( up_outs )
pdf_fitted = stats.expon.pdf( time[:400], loc = param[0], scale = param[1] )
plot( time[:400], pdf_fitted, 'r-' )
pdf_fitted = stats.expon.pdf( time, loc = param[0], scale = 99.46 )
plot( time, pdf_fitted, 'g-' )
xlabel( 'Time (s)', fontsize=20 )
ylabel( 'Density ($s^{-1}$)', fontsize=20 )
plt.hist(U_O, 200, histtype = 'stepfilled', normed = True)
plt.text(0.2,.6,'$\\tau_{\uparrow ,out}=$132$\pm$66s' % param[1], fontsize = 20)
xlim([0.05,0.45])
ylim([0,1])
show()
```
