```python
#Add Gaussian noise to the waveform.
#Execute Gaussian transformation to each point of the waveform.

points = 1000
deltat = 0.018 #s, taken from spin to charge optimisation

t_down_in = 0.0039
t_up_out = 0.0069 #Taken from experimental data.
deltaI = 32 * 10 ** - 12 #A

time = linspace( 0, deltat, points )
wave = time[ : ] * 0

start = int( ( t_up_out / deltat ) * points )
finish = int( ( ( t_up_out + t_down_in ) / deltat ) * points )

wave[ start: finish ] =  ( time[ start: finish ] * 0 ) + deltaI 

a_n = 71 * 10 ** - 15 #A / ( ( Hz ) ^ ( 1 / 2 ) ) #Have to multiply spectral density by a bandwidth.
f_c = 300 #Hz, cut off frequency, also used as the bandwidth
A_n = a_n * sqrt( f_c )

z1 = linspace( -1, 1, points )

p = exp( ( - z1 ** 2 ) / ( 2 * ( A_n ** 2 ) ) ) / ( A_n * sqrt( 2 * pi ) )

figure()
plot( z1, p )

figure()
plot( time, wave )

noise = random.normal( 0, A_n, points )

z2 = z1 * 0
z3 = z2 + noise

figure()
plot( z1, z3 )

noisy_wave = wave + noise

figure()
plot( time, noisy_wave)
#Successfully adds noise to the signal.
#With the given numbers the noise seems insignificant compared to the signal.
```

```python
import scipy.signal as sig

#a function to filter a trace
def lowpass(time,data,cutoff):
    samprate = 1.0/(time[1]-time[0])
    b,a = sig.bessel(8,cutoff/(samprate/2.0),btype='low',analog=0,output='ba')
    #note cutoff is given relative to nyquist(=sample rate/2)
    data_f = sig.filtfilt(b,a,data)
    return data_f
```

```python
#Apply the filter to the noisy waveform.
filtered_wave = lowpass( time, noisy_wave, 300 )

figure()
plot( time, filtered_wave, 'b', time, noisy_wave, 'gx' )
```
