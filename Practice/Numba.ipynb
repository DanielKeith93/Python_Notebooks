```python
from numba import jit
from numpy import arange

# jit decorator tells Numba to compile this function.
# The argument types will be inferred by Numba when function is called.
@jit
def sum2d(arr):
    M, N = arr.shape
    result = 0.0
    for i in range(M):
        for j in range(N):
            result += arr[i,j]
    return result

a = arange(9).reshape(3,3)
print (sum2d(a))
```

```python
from scipy import signal

#define parameters

fs = 20000.0 #Hz sample rate
lowcut = 10000.0 #Hz filter cutoff frequency
T = 0.055 #s readout time
nsamples = T * fs #number of sample points
N_0 = 0.55 #spin up fraction
N_1 = 1 - N_0
deltat = T #s
t_down_in = 0.0051 #s characteristic tunnel times
t_up_out = 0.0065 
deltaI = 0.45 * 10 ** - 9 #A signal height
points = nsamples
time1 = linspace( 0, deltat, points ) #time axis for traces
a_n = 0.24 * 10 ** - 12 #A / ( ( Hz ) ^ ( 1 / 2 ) ) #Have to multiply spectral density by a bandwidth.
f_c = lowcut #Hz, cut off frequency, also used as the bandwidth
A_n = a_n * sqrt( fs ) #A dependence on the sample rate
bars = 200 #numbers of bars for the histogram
traces = 10000 #number of traces to simulate
```

```python
from scipy.signal import bessel, lfilter

#filter functions that do not work properly
#more recent notebooks have a filter function that works

@jit
def bessel_lowpass(lowcut,fs, order=8):
    nyq = 0.5 * fs
    low = lowcut / nyq
    b, a = bessel(order, low, btype='low')
    return b, a

@jit
def bessel_lowpass_filter(data, lowcut,fs, order=8):
    b, a = bessel_lowpass(lowcut,fs, order=order)
    y = lfilter(b, a, data)
    return y

# a function to make square pulses

@jit
def sqr_pulse( c, time_arr, t1, t2, I ):
    start = int( ( t1 / deltat ) * len(time_arr) )
    finish = int( ( ( t1 + t2 ) / deltat ) * len(time_arr) )
    if c == 0:
        wave = time_arr[ : ] * 0
        wave[ start: finish ] =  ( time_arr[ start: finish ] * 0 ) + I 
        return wave
    elif c == 1:
        wave = time_arr[ : ] * 0
        return wave

#wavelet used in the continuous wavelet transform
    
@jit    
def DOG(a, length):
    t = np.linspace(-length/2, (length)/2, length)
    y = - a ** 2 * t * exp( - t ** 2 / ( 2 * a ** 2 ) ) #/ ( sqrt( 2 * pi ) * a ** 3 )
    return y

#can smooth out traces to get rid of small unimportant peaks
#may not work
#isn't necessary for calculations to work

@jit
def smooth(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='same')
    return y_smooth
```

```python
#calculates the maximum weight of a trace

@jit
def edge_max( z ):
    widths = np.arange(1, 101)
    cwtmatr = signal.cwt(z, DOG, widths) #performs continuous wavelet transform
    peaks = []
    for i in cwtmatr:    
        max_min = []
        if array_equal(cwtmatr[-1],i) == True: #only for last scale of transform
            b = smooth( i, 9 ) #smooths out the transform at each scale, may cause problems
            d = abs( b )
            for x in xrange( 1, 1099 ):
                if b[x-1]>b[x] and b[x+1]>b[x]: #search for peaks in transform
                    if b[x]<-median(d):
                        max_min.append( x )
                if b[x-1]<b[x] and b[x+1]<b[x]:
                    if b[x]>median(d):
                        max_min.append( x )
            peaks.append( max_min )
        else: #for all other scales   
            b = smooth( i, 9 )
            for x in xrange( 1, 1099 ):
                if b[x-1]>b[x] and b[x+1]>b[x]:
                    max_min.append( x )
                if b[x-1]<b[x] and b[x+1]<b[x]:
                    max_min.append( x )
            peaks.append( max_min )
    tracks = []
    
    for i in peaks[-1]: #find a track for each peak in the last scale of the transform
        track = []
        x = i
        track.append( x )
        for s in xrange( 1, 100 ): #find the closest peak at the next scale
            a = peaks[ - s - 1 ]
            if x < a[-1]:
                X = np.searchsorted(a, x, side="left")
                if math.fabs(x - a[X-1]) < math.fabs(x - a[X]):
                    idx = X-1
                else:
                    idx = X
            else:
                idx = len(a) - 1
            x = peaks[ - s - 1 ][ idx ]
            track.append( x )
        tracks.append( track )
    norm = []
    y = []
    for l in xrange( 0, 100 ): #calculate values to normalise the transform for each scale
        for k in cwtmatr[ l ]:
            a = k ** 2
            y.append( a )
        med = median( y )
        norm.append( med )
    weights = []
    for i in xrange( 0, len( peaks[-1] ) ): #sum up the normalised values of the transform along the entire track for each peak
        weight = []
        for j in tracks[ i ]:
            scale = 99
            a = cwtmatr[ scale ][ j ] ** 2
            b = norm[ scale ]
            weight.append( a / b )
            scale -= 1
        c = sum( weight )    
        weights.append( c )
    return max( weights ) #returns the maximum weight from all the tracks within a trace
```

```python
#finds the point closest to another, arbitrary point

def find_nearest(array,value):
    idx = np.searchsorted(array, value, side="left")
    if math.fabs(value - array[idx-1]) < math.fabs(value - array[idx]):
        return array[idx-1]
    else:
        return array[idx]
```

```python
#generates random traces and calculate the maximum weight of each one

@jit
def gen_wav4( n ):
    N_up = []
    N_down = []
    for i in xrange( 0, n):
        if randint(0,99) < ( N_0 * 100 ):
            charge = 0
        else:
            charge = 1
        t_in = t_down_in 
        t_out = t_up_out
        r_in = randint( 1,1001 )
        r_out = randint( 1,1001 )
        t_in = - log( r_in * 0.001 ) * t_in
        t_out = - log( r_out * 0.001 ) * t_out
        sig = bessel_lowpass_filter(sqr_pulse( charge, time1, t_out, t_in, deltaI ) + A_n * random.normal( 0, 1, points ), lowcut,fs, order=8)
        a = edge_max( sig )
        if charge == 0:
            N_up.append( a )
        else:
            N_down.append( a )

    return N_up, N_down
```
