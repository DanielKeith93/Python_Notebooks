```python
#Setup parameters
N_0 = 0.37
N_1 = 1 - N_0 #These numbers were chosen to best fit experimental data

deltat = 0.018 #s, taken from spin to charge optimisation
t_down_in = 0.0039
t_up_out = 0.0069 #Taken from experimental data.
deltaI = 32e-12 #A (32pA)
```

```python
# a function to make square pulses
def sqr_pulse(charge, time_arr, t1, t2, deltaI):
    '''sqr_pulse(charge, time_arr, t1, t2)
    IN: charge = int, 0 or 1
        time_arr = array-like of floats defining time-points
        t1 = delay (pulse switches on)
        t2 = duration (between on and off)
        deltaI = peak current
    OUT: array-like pulse matching the size of time_arr
    '''
    wave = time[ : ] * 0
    start = int( ( t1 / deltat ) * len(time) )
    finish = int( ( ( t1 + t2 ) / deltat ) * len(time) )
    peak = ( time[ : ] * 0 ) 
    peak[ start: finish ] =  ( time[ start: finish ] * 0 ) + deltaI 
    
    if charge == 0:
        wave = peak
    return wave
```

```python
#Generate the particular waveform depending on the charge input, first '0'

charge = 0 #Charge 0 is supposed to return a peak in the waveform.
points = 1000

time = linspace( 0, deltat, points )
wave = sqr_pulse(0, time, t_up_out, t_down_in, deltaI)
wave2 = sqr_pulse(1, time, t_up_out, t_down_in, deltaI)
    
figure()
plot( time, wave,'b-', time , wave2, 'rx')

#Generates two different waves depending on the charge input.
```

```python
import scipy.signal as sig

#a function to filter a trace
def lowpass(time,data,cutoff):
    samprate = 1.0/(time[1]-time[0])
    b,a = sig.butter(2,cutoff/(samprate/2.0),btype='low',analog=0,output='ba')
    #note cutoff is given relative to nyquist(=sample rate/2)
    data_f = sig.filtfilt(b,a,data)
    return data_f
```
