```python
#define parameters

fs = 20000.0 #Hz sample rate
lowcut = 10000.0 #Hz filter cutoff frequency
T = 0.055 #s readout time
nsamples = T * fs #number of samplep oints
N_0 = 0.55 #spin up fraction
N_1 = 1 - N_0
deltat = T #s
t_down_in = 0.0051 #characteristic tunnel times
t_up_out = 0.0065 
deltaI = 0.45 * 10 ** - 9 #A signal height
points = nsamples
time1 = linspace( 0, deltat, points ) #time axis for traces
a_n = 2.4 * 10 ** - 12 #A / ( ( Hz ) ^ ( 1 / 2 ) ) #Have to multiply spectral density by a bandwidth.
f_c = lowcut #Hz, cut off frequency, also used as the bandwidth
A_n = a_n * sqrt( fs ) #A dependent on sample rate
bars = 200 #bars in histogram
traces = 10000 #traces to simulate
```

```python
from scipy.signal import bessel, lfilter

#filter functions that don't work properly
#more recent notebooks include filter function that works

def bessel_lowpass(lowcut,fs, order=8):
    nyq = 0.5 * fs
    low = lowcut / nyq
    b, a = bessel(order, low, btype='low')
    return b, a


def bessel_lowpass_filter(data, lowcut,fs, order=8):
    b, a = bessel_lowpass(lowcut,fs, order=order)
    y = lfilter(b, a, data)
    return y

#generates the ideal shape of a simulated trace with specified dimensions

def sqr_pulse( c, time_arr, t1, t2, I ):
    start = int( ( t1 / deltat ) * len(time_arr) )
    finish = int( ( ( t1 + t2 ) / deltat ) * len(time_arr) )
    if c == 0:
        wave = time_arr[ : ] * 0
        wave[ start: finish ] =  ( time_arr[ start: finish ] * 0 ) + I 
        return wave
    elif c == 1:
        wave = time_arr[ : ] * 0
        return wave
```

```python
#creat a simulated trace

z = bessel_lowpass_filter(sqr_pulse( 0, time1, t_up_out, t_down_in, deltaI ) + A_n * random.normal( 0, 1, points ), lowcut,fs, order=8)
z2 = sqr_pulse( 0, time1, t_up_out, t_down_in, deltaI )
```

```python
#derivative of a gaussian function to use as transform wavelet

def DOG(a, length):
    t = np.linspace(-length/2, (length)/2, length)
    y = - a ** 2 * t * exp( - t ** 2 / ( 2 * a ** 2 ) ) #/ ( sqrt( 2 * pi ) * a ** 3 )
    return y
```

```python
#continuous wavelet transform of a simulated trace using DOG wavelet

t = np.linspace(-1, 1, 200, endpoint=False)
sig  = z
widths = np.arange(1, 101)
cwtmatr = signal.cwt(sig, DOG, widths)
fig = plt.imshow(cwtmatr, extent=[0, 0.055, 1, 100], cmap='afmhot', aspect='auto', vmax=abs(cwtmatr).max(), vmin=-abs(cwtmatr).max())
ylabel( "Wavelet Scale $s$", fontsize=18 )
#plt.yscale( 'log' )
xlabel( "Time (s)", fontsize=18 )
cbar = plt.colorbar()
cbar.ax.set_ylabel( "$W(t,s)$", fontsize=18 )
savefig('C:\Users\z3414503\Desktop\Honours\LitReview\Pictures\Cwt2.png', format='png', dpi=600)

plt.show()
```

```python
#find all the maxima and minima in a transform

peaks = []
for i in cwtmatr: 
    max_min = []
    b = smooth( i, 100 )
    for x in xrange( 1, 1099 ):
        if b[x-1]>b[x] and b[x+1]>b[x]:
            max_min.append( x )
        if b[x-1]<b[x] and b[x+1]<b[x]:
            max_min.append( x )
    peaks.append( max_min )
```

```python
#print out the peaks

print peaks[-1]
print len( peaks[ -1 ] )
```

```python
#calculate the track for the first peak
#uses code that causes problems for numba
#this has been replaced in more recent notebooks

track = []
x = peaks[ -1 ][ 0 ]
track.append( x )
for s in xrange( 1, 100 ):
    a = peaks[ - s - 1 ]
    idx = min(range(len(a)), key=lambda i: abs(a[i]-x)) #line that numba can't understand
    x = peaks[ - s - 1 ][ idx ]
    track.append( x )
print track
print len( track )
```

```python
#calculate the tracks of all the peaks

tracks = []
for i in peaks[ -1 ]:
    track = []
    x = i
    track.append( x )
    for s in xrange( 1, 100 ):
        a = peaks[ - s - 1 ]
        idx = min(range(len(a)), key=lambda i: abs(a[i]-x))
        x = peaks[ - s - 1 ][ idx ]
        track.append( x )
    tracks.append( track )
print len( tracks )
```

```python
#normalisation constants for each of the scales

norm = []
y = []
for l in xrange( 0, 100 ): #each scale
    for k in cwtmatr[ l ]:
        a = k ** 2
        y.append( a )
    med = median( y ) #median of the squares of the transform at each scale
    norm.append( med )
print norm
print len( norm )
```

```python
#calculate weight by summing up all normalised transform values along a track

weights = []
for i in xrange( 0, len( peaks[ -1 ] ) ):
    weight = []
    for j in tracks[ i ]:
        scale = 99
        a = cwtmatr[ scale ][ j ] ** 2
        b = norm[ scale ]
        weight.append( a / b )
        scale -= 1
    c = sum( weight )    
    weights.append( c )
```

```python
#code from internet that takes an image and outputs another image with just the edges of the inout image

import scipy.ndimage as ndi
import scipy
import numpy
import Image
import math
sigma = 1.4
f = 'C:/Users/z3414503/Desktop/lion.jpg'
img = Image.open(f).convert('L') #grayscale
imgdata = numpy.array(img, dtype = float)
G = ndi.filters.gaussian_filter(imgdata, sigma)
sobelout = Image.new('L', img.size)                                       #empty image
gradx = numpy.array(sobelout, dtype = float)                        
grady = numpy.array(sobelout, dtype = float)
 
sobel_x = [[-1,0,1],
           [-2,0,2],
           [-1,0,1]]
sobel_y = [[-1,-2,-1],
           [0,0,0],
           [1,2,1]]
 
width = img.size[1]
height = img.size[0]
 
for x in range(1, width-1):
    for y in range(1, height-1):
        px = (sobel_x[0][0] * G[x-1][y-1]) + (sobel_x[0][1] * G[x][y-1]) + \
             (sobel_x[0][2] * G[x+1][y-1]) + (sobel_x[1][0] * G[x-1][y]) + \
             (sobel_x[1][1] * G[x][y]) + (sobel_x[1][2] * G[x+1][y]) + \
             (sobel_x[2][0] * G[x-1][y+1]) + (sobel_x[2][1] * G[x][y+1]) + \
             (sobel_x[2][2] * G[x+1][y+1])
 
        py = (sobel_y[0][0] * G[x-1][y-1]) + (sobel_y[0][1] * G[x][y-1]) + \
             (sobel_y[0][2] * G[x+1][y-1]) + (sobel_y[1][0] * G[x-1][y]) + \
             (sobel_y[1][1] * G[x][y]) + (sobel_y[1][2] * G[x+1][y]) + \
             (sobel_y[2][0] * G[x-1][y+1]) + (sobel_y[2][1] * G[x][y+1]) + \
             (sobel_y[2][2] * G[x+1][y+1])
        gradx[x][y] = px
        grady[x][y] = py
sobeloutmag = scipy.hypot(gradx, grady)
sobeloutdir = scipy.arctan2(grady, gradx)
for x in range(width):
    for y in range(height):
        if (sobeloutdir[x][y]<22.5 and sobeloutdir[x][y]>=0) or \
           (sobeloutdir[x][y]>=157.5 and sobeloutdir[x][y]<202.5) or \
           (sobeloutdir[x][y]>=337.5 and sobeloutdir[x][y]<=360):
            sobeloutdir[x][y]=0
        elif (sobeloutdir[x][y]>=22.5 and sobeloutdir[x][y]<67.5) or \
             (sobeloutdir[x][y]>=202.5 and sobeloutdir[x][y]<247.5):
            sobeloutdir[x][y]=45
        elif (sobeloutdir[x][y]>=67.5 and sobeloutdir[x][y]<112.5)or \
             (sobeloutdir[x][y]>=247.5 and sobeloutdir[x][y]<292.5):
            sobeloutdir[x][y]=90
        else:
            sobeloutdir[x][y]=135
mag_sup = sobeloutmag
for x in range(1, width-1):
    for y in range(1, height-1):
        if sobeloutdir[x][y]==0:
            if (sobeloutmag[x][y]<=sobeloutmag[x][y+1]) or \
               (sobeloutmag[x][y]<=sobeloutmag[x][y-1]):
                mag_sup[x][y]=0
        elif sobeloutdir[x][y]==45:
            if (sobeloutmag[x][y]<=sobeloutmag[x-1][y+1]) or \
               (sobeloutmag[x][y]<=sobeloutmag[x+1][y-1]):
                mag_sup[x][y]=0
        elif sobeloutdir[x][y]==90:
            if (sobeloutmag[x][y]<=sobeloutmag[x+1][y]) or \
               (sobeloutmag[x][y]<=sobeloutmag[x-1][y]):
                mag_sup[x][y]=0
        else:
            if (sobeloutmag[x][y]<=sobeloutmag[x+1][y+1]) or \
               (sobeloutmag[x][y]<=sobeloutmag[x-1][y-1]):
                mag_sup[x][y]=0
m = numpy.max(mag_sup)
th = 0.2*m
tl = 0.1*m
 
 
gnh = numpy.zeros((width, height))
gnl = numpy.zeros((width, height))
 
for x in range(width):
    for y in range(height):
        if mag_sup[x][y]>=th:
            gnh[x][y]=mag_sup[x][y]
        if mag_sup[x][y]>=tl:
            gnl[x][y]=mag_sup[x][y]
gnl = gnl-gnh
 
def traverse(i, j):
    x = [-1, 0, 1, -1, 1, -1, 0, 1]
    y = [-1, -1, -1, 0, 0, 1, 1, 1]
    for k in range(8):
        if gnh[i+x[k]][j+y[k]]==0 and gnl[i+x[k]][j+y[k]]!=0:
            gnh[i+x[k]][j+y[k]]=1
            traverse(i+x[k], j+y[k])
 
for i in range(1, width-1):
    for j in range(1, height-1):
        if gnh[i][j]:
            gnh[i][j]=1
            traverse(i, j)
```
